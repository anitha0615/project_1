```python
url = "https://data.texas.gov/resource/mwzi-gyw7.json"

# Set up pagination parameters
offset = 0
limit = 1000

results = []

while True:
    # Set up the API request with pagination parameters
    params = {'$offset': offset, '$limit': limit}
    response = requests.get(url, params=params)
    
    # Create a DataFrame from the API response
    df = pd.read_json(response.text)
    
    # Append the DataFrame to the overall results list
    results.append(df)
    
    # Check if there are more results
    if len(df) < limit:
        break
    
    # Increment the offset for the next request
    offset += limit

# Concatenate all the DataFrames into one
df_all = pd.concat(results, ignore_index=True)

# Print the final DataFrame
print(df_all)
```

```python
print(f"We received {len(df_all)} responses.")
```

```python
# drop the rows with blank values in the "zip_cd" column
df1 = df_all.dropna(subset=['zip_cd'])
```

```python
print(f"We received {len(df1)} responses.")
```

```python
df1 = pd.DataFrame(df1, columns=['zip_cd', 'phys_loc', 
            'violation_category', 'media_name', 
            'violation_status', 'rule_citation', 
            'investigation_no',"allegation_txt"])
df1
```
